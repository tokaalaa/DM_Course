```
import numpy as np
```

```
Data_Matrix = np.array([[ 10., 60.,  10., 90.],
                        [ 20., 50.,  40., 70.],
                        [ 30., 50.,  30., 40.],
                        [ 20., 50.,  20., 60.],
                        [ 10., 60.,  30., 10.]])
```

```
# Data Matrix Q2
#Compute the norm of each instance. (5x1)
print(np.linalg.norm(Data_Matrix, axis=1))

#Compute the Cosine similarity matrix (5x5) matrix
d = Data_Matrix @ Data_Matrix.T
norm = (Data_Matrix * Data_Matrix).sum(1, keepdims=True) ** .5
Cosine_similarity = d / (norm * norm.T)
print(Cosine_similarity)

#Compute the Euclidean Distance matrix of the instances (5x5)
##(a-b)**.5 = a.a + b.b - 2ab
n,d = Data_Matrix.shape
G = np.dot(Data_Matrix,Data_Matrix.T)
H = np.tile(np.diag(G),(n,1))
Euclidean_Distance = H + H.T - 2*G
print(Euclidean_Distance)
```

```
Data_Matrix = np.array([[ 10., 60., 90.],
                        [ 20., 50., 70.],
                        [ 30., 50., 40.],
                        [ 20., 50., 60.],
                        [ 10., 60., 10.]])
```

```
# Import libraries
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
 
# Creating figure
fig = plt.figure(figsize = (8, 8))
ax = plt.axes(projection ="3d")
 
# Creating plot
ax.scatter3D(Data_Matrix[:,0].T,Data_Matrix[:,1],Data_Matrix[:,2], color = "green")
ax.set_xlabel('X-axis', fontweight ='bold') 
ax.set_ylabel('Y-axis', fontweight ='bold') 
ax.set_zlabel('Z-axis', fontweight ='bold')

# show plot
plt.show()
```

```
#Compute the mean vector (3x1)
mean = Data_Matrix.mean(axis=0)

#Compute centered data matrix Z by subtracting mean vector from the Data Matrix. (5x3)
Z = Data_Matrix - mean
print(Z)
n,d = Z.shape

#Compute Covariance matrix COV (3x3)
cov = Z.T@ Z/n
print("cov = ",cov)
#Compute eigvals & eigvecs
y,U = np.linalg.eigh(cov)
print("y = ",y)
print("U = ",U)

#Verify UTΛU= COV
print(U @ np.diag(y) @ U.T)

# total variance
var = sum(y)
print("total variance= ",var)
print("fraction of total variance for the largest eigval= ", y[2]/var)
```

```
#Compute the projection matrix P to go to 2-dimensions. Consider
#the top two eigenvectors of matrix U according to
#eigenvalues.(3x2)
P = U[:,[1,2]]
print(P)

#Project the instances into a 2-Dimension space. xນ= PTx
print(Data_Matrix.shape, P.shape)
D2 = Data_Matrix @ P
print(D2)
```

```
import matplotlib.pyplot as plt

plt.scatter(D2[:,0], D2[:,1])
plt.show()
```
